## Удаление ячеек в таблице

```python
df.at[3, 'Класс'] = None
df.at[1, 'Вес'] = None
df.at[39, 'Рост'] = None
```

```python
df
```

## Удаление строк с пропущенными значениями

```python
df.dropna()
```

## Заполнение пропущенных значений средним значением по колонке

```python
df.fillna(df.mean(), inplace=True)
```

## Создание нового поля через выражение на базе имеющихся колонок

```python
df['Рост + вес'] = df['Рост'] + df['Вес']
```

```python
df
```

## Создание нового поля через DataFrame.apply

```python
df['Рост/Вес'] = df.apply(lambda row: row['Рост'] / row['Вес'], axis=1)
```

```python
df
```

## Создание нового поля через Series.apply

```python
df['Двойной вес'] = df['Вес'].apply(lambda x: x * 2)
```

```python
df
```

## Сортировка по полю

```python
sorted_df = df.sort_values('Рост', ascending=False)
sorted_df
```

## Вычисление статистик по колонкам

```python
mean_value = df['Рост'].mean()
max_value = df['Рост'].max()
min_value = df['Рост'].min()
count_five = df.query('Класс == 5').count()[0]
print(mean_value)
print(max_value)
print(min_value)
print(count_five)
```

## Использование .value_counts()

```python
value_counts = df['Вес'].value_counts()
value_counts
```

## Вывод уникальных значений колонки

```python
unique_values = df['Имя'].unique()
unique_values
```

## Удаление текущего индекса и создание нового на базе новой колонки

```python
df.reset_index(drop=True, inplace=True)
df.set_index('Двойной вес', inplace=True)
```
